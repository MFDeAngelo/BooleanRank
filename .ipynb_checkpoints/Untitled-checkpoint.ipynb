```python
VERTICES = 11
tournaments = digraphs.tournaments_nauty(VERTICES,VERTICES // 2,VERTICES // 2)
fileOutput = file('output/minCliqueNumbers.txt','w')



def main():
    index = 1
    for g in tournaments:
        doIt(g)
        if index % 100 == 0:
            print index
        index = index + 1
    print "Done!"



def testOne():
    graph = 0
    for g in tournaments:
        graph = g
    adjacencyMatrix = graph.adjacency_matrix()
    bipartiteCliques = getBipartiteCliques(adjacencyMatrix)
    arcList = getArcList(graph, adjacencyMatrix)
    bicliqueMatrix = getBicliqueMatrix(arcList, bipartiteCliques)
    solveLinearProgram(bicliqueMatrix)

 
############################################## Cliques


def doIt(T):
    adjacencyMatrix = T.adjacency_matrix()
    bipartiteCliques = getBipartiteCliques(adjacencyMatrix)
    arcList = getArcList(T, adjacencyMatrix)
    bicliqueMatrix = getBicliqueMatrix(arcList, bipartiteCliques)
    minCliqueNumber = solveLinearProgram(bicliqueMatrix)
    fileOutput.write(str(adjacencyMatrix) + "\nMinimum Clique Number: " + str(minCliqueNumber) +"\n\n")
    
    
    

def solveLinearProgram(bicliqueMatrix):
    lp = MixedIntegerLinearProgram(maximization=False)
    x = lp.new_variable(real=True, nonnegative=True)
    solVector = [0] * bicliqueMatrix.nrows()
    for i in range(bicliqueMatrix.nrows()):
        for j in range(bicliqueMatrix.ncols()):
            solVector[i] = bicliqueMatrix[i,j]*x[j] + solVector[i]
    sigmaX = 0
    for i in range(bicliqueMatrix.ncols()):
        lp.set_max(x[i],1)
        sigmaX = sigmaX + x[i]
    lp.set_objective(sigmaX)
    for xs in solVector:
        lp.add_constraint(xs >= 1)
    return lp.solve()
    
        

def getBicliqueMatrix(arcList, bipartiteCliques):
    columns = []
    for clique in bipartiteCliques:
        column = [0] * len(arcList)
        for arc in clique:
            column[arcList.index(arc)] = 1
        columns.append(column)
    return matrix(columns).transpose()


def getArcList(tournament, bicliqueList):
    arcList = []
    for vertex1 in tournament.vertices():
        for vertex2 in tournament.neighbors_out(vertex1):
            arc = [vertex1, vertex2]
            arcList.append(arc)
    return arcList
            

def getBipartiteCliques(matrix): #Only works for regular tournaments
    R = matrix.nrows()
    bicliqueList = []
    for i in range(1, VERTICES //2 + 1):
        for combination in Combinations(range(R), i).list():
            clique = getBipartiteClique(matrix, combination)
            if clique != []:
                bicliqueList.append(clique)
    return bicliqueList


def getBipartiteClique(matrix, sourceList):
    clique = []
    for i in range(matrix.ncols()):
        addVertex = true
        for source in sourceList:
            if(matrix[source,i] == 0):
                addVertex = false
        if(addVertex):
            for source in sourceList:
                clique.append([source,i])
    return clique



############################################# Matrix Opperations

def matrixMultiply(matrix1, matrix2):
    results = matrix1 * matrix2
    for i in range(VERTICES):
        for j in range(VERTICES):
            if(results[i,j] > 1):
                results[i,j] = 1
    return results
    
    
def addMultiply(matrix1, matrix2): #This can be sped up...
    results = matrix1 + matrix2
    for i in range(VERTICES):
        for j in range(VERTICES):
            if(results[i,j] > 1):
                results[i,j] = 1
    return results
```
